```python
import pandas as pd
import os

project_path = os.path.dirname(os.path.dirname(os.getcwd()))
data_path = os.path.join(project_path, "assets/megaGymDataset.csv")
data_path
```

```python
df = pd.read_csv(data_path)
df.dropna(inplace=True)
df.drop(['Unnamed: 0','Rating', 'RatingDesc'], axis=1, inplace=True)
# df.reset_index(drop=True, inplace=True)
df[:500]
```

```python
df.iloc[0]['Title']
```

```python
# convert df to text format
with open(os.path.join(project_path, "assets/megaGymDataset.txt"), "w+") as file:
    for i in range(len(df)):
        title = df.iloc[i]['Title']
        desc = df.iloc[i]['Desc']
        type = df.iloc[i]['Type']
        bodypart = df.iloc[i]['BodyPart']
        equip = df.iloc[i]['Equipment']
        level = df.iloc[i]['Level']
        string = f"Exercise name: {title}\nExercise Description: {desc}\nExercise Type: {type}\nExercise BodyPart: {bodypart}\nExercise Equipment: {equip}\nExercise Level: {level}\n----------+\n"
        file.write(string)



```

```python
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter, CharacterTextSplitter

loader = TextLoader(os.path.join(project_path, "assets/megaGymDataset.txt"))
text_documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size = 400, chunk_overlap=20, separator='\n----------+\n', strip_whitespace=False)
text_splitter.split_documents(text_documents)[:5]
```
